```python
from transformers import BertTokenizer
from tqdm import trange, tqdm

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
print(len(data.known_label_list))
```

```python
test_dataloader = data.dataloader.test_loader
```

```python
total_labels = torch.empty(0,dtype=torch.long).to(method.device)
total_preds = torch.empty(0,dtype=torch.long).to(method.device)

total_features = torch.empty((0,args.feat_dim)).to(method.device)
```

```python
from utils.functions import save_model, euclidean_metric
import torch.nn.functional as F

def open_classify(data, features):
    logits = euclidean_metric(features, method.centroids)
    print(logits[0])
    probs, preds = F.softmax(logits.detach(), dim = 1).max(dim = 1)
    euc_dis = torch.norm(features - method.centroids[preds], 2, 1).view(-1)
    preds[euc_dis >= method.delta[preds]] = data.unseen_label_id

    return preds
```

```python
for batch in tqdm(test_dataloader):
    batch = tuple(t.to(method.device) for t in batch)
    
    input_ids, input_mask, segment_ids, label_ids = batch
    print()
    print(input_ids.shape)
    print(tokenizer.batch_decode(input_ids))
    
    with torch.set_grad_enabled(False):
        pooled_output = method.model(input_ids, segment_ids, input_mask, feature_ext=True)
        print(pooled_output.shape)
        
        preds = open_classify(data, pooled_output)
        print(preds)
        
        total_preds = torch.cat((total_preds, preds))
        total_labels = torch.cat((total_labels, label_ids))
        total_features = torch.cat((total_features, pooled_output))
        
    break
```
